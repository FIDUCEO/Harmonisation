## 1.  The Gridding Process

A simple process of gridding a set of pixels, ${z_i}$, is the following,

$\langle z \rangle = \frac{1}{N} \sum_i^N z_i$

The uncertainty of this process may be defined as,

$u^2(\langle z \rangle) = u_{\mathrm{ind.}}^2(\langle z \rangle) + u_{\mathrm{ind.}}^2(\langle z \rangle)$

where:

* $u_{\mathrm{ind.}}^2(\langle z \rangle)$ - is the uncertainty of the grid cell due to independent error effects
* $u_{\mathrm{com.}}^2(\langle z \rangle)$ - is the uncertainty of the grid cell due to common error effects
* Note that uncertainty due to structured error effects are omitted

In this exercise we will look at gridding the FIDUCEO easyFCDR AVHRR sample file to a L3 grid, propagating the independent and common uncertainty components through the processing.

## 2. Uncertainties in Gridding

### ``i. Independent Uncertainties``

In an image, independent errors therefore arise from random effects operating at the image pixel level, the classic example being detector noise. The uncertainty of the result of an averaging process for these kind of effects is:

$$ u_{\mathrm{ind.}}^2(\langle z \rangle) = \frac{1}{N^2}\sum_i^N u_{\mathrm{ind.}}^2(z_i)$$

Therefore the contribution of the independent uncertainty to the combined uncertainty  $u_c(\langle z \rangle )$ will follow the relationship:

$$ u_{\mathrm{ind.}}^2(\langle z \rangle) \propto \frac{1}{\sqrt{N}}$$

```python
IndependentUFig(N_values,x_val,uncertainties,cbar_range,cmap,Names,Titlestart)
```

### ``ii. Common Uncertainties``

Common errors are common between satellite pixels for an entire mission. The uncertainty of the result of an averaging process for these kind of effects is:

$$ u_{\mathrm{com.}}^2(\langle z \rangle) = \frac{1}{N^2}\left [ \sum_i^N u_{\mathrm{com.}}(z_i) \right ]^2 = \langle u_{\mathrm{com.}}(z_i) \rangle^2$$

We find therefore that this type of uncertainty does not reduce with N.

```python
CommonUFig(N_values,x_val,uncertainties,cbar_range,cmap,Names,Titlestart)
```

## 3. Interactive Comparison Using Real Data

The graphs below illustrate the combination of independent and common uncertainties for a grid cell in the centre of our AVHRR easyFCDR sample data.

The sliders allow you to change:

* The fraction of uncertainty coming from independent and common error sources
* The size of the grid cell
* The X position of the grid cell
* The Y position 

```python
ThreeErrors(uncertainties,cbar_range,cmap,Titlestart,Names,v1=False)
@widgets.interact(Percentage=(0,1,0.1),Size=(2,100,5),X=(0,400,10),Y=(0,700,20))
def P(Percentage,Size,X,Y):
    fig = plt.figure(figsize=(6,8))
    param = Percentage
    Size = Size
    WindowU = []
    ModI = []
    for elem in uncertainties:
        W , Mo = Scan(elem,np.array([Size,Size]),np.array([X,Y]))
        WindowU.append(W)
        ModI.append(Mo)
    WindowU = np.array(WindowU)
    ModI = np.array(ModI)
    Im = np.array([WindowU,ModI])
    BackU = CombinedDa(Im,param,Size,Averaging=False,ShowStructured=False)
    WindowU = CombinedDa(Im,param,Size,Averaging=True,ShowStructured=False)
    plt.imshow(BackU + WindowU,cmap=cmap,vmin=cbar_range[0],vmax=cbar_range[1]) 
    plt.title("Combined Uncertainties",size=16)
    plt.colorbar(shrink=0.9,aspect=50),#orientation='horizontal')
    plt.tick_params(axis='both', left='off', top='off', right='off', bottom='off', labelleft='off', labeltop='off', labelright='off', labelbottom='off')
```

## 4. Extension Exercises

The interactive plot below shows a zoomed-in version of the previous graph. The idea is to show how both the gridding size and the proportion of common to independent uncertanty play a role in the value for the combined standard uncertainty. Here, the figure axis scale varies with the window size in order to better represent changes to the combined standard uncertainty. The graph on the right plots the value of the combined standard uncertainty and will update everytime there is a change to one of the sliders. Here Slide refers to the side length of the averaging window, such that $N=S^2$. Keep this in mind when observing the *$\frac{1}{\sqrt{N}}$* behaviour , as you will be seeing *$\frac{1}{S}$* behaviour.

```python
window_values = []
@widgets.interact(Percentage=(0,1,0.01),Size=(1,30,1),X=(0,400,10),Y=(0,700,20))
def P(Percentage,Size,X,Y):
    plt.figure(figsize=(10,7))
    cmap = "gnuplot2"
    cbar_range = np.array([0,0.16])
    param = Percentage
    Size = Size
    WindowU = []
    ModI = []
    for elem in uncertainties:
        W , Mo = Scan(elem,np.array([Size,Size]),np.array([X,Y]))
        WindowU.append(W)
        ModI.append(Mo)
    WindowU = np.array(WindowU)
    ModI = np.array(ModI)
    Im = np.array([WindowU,ModI])
    BackU = CombinedDa(Im,param,Size,Averaging=False,ShowStructured=False)
    WindowU = CombinedDa(Im,param,Size,Averaging=True,ShowStructured=False)
    plt.subplot(121)
    if Y-4*Size>0 and X-3*Size>0:
        plt.imshow(BackU[Y-4*Size:Y+4*Size,X-3*Size:X+3*Size]+WindowU[Y-4*Size:Y+4*Size,X-3*Size:X+3*Size],cmap=cmap,vmin=cbar_range[0],vmax=cbar_range[1])
    else:
        plt.imshow(BackU[0:Y+4*Size,0:X+3*Size]+WindowU[0:Y+4*Size,0:X+3*Size],cmap=cmap,vmin=cbar_range[0],vmax=cbar_range[1])
    Windowvalue = WindowU[np.nonzero(WindowU)][0]
    window_values.append(Windowvalue)
    plt.subplot(121)
    plt.colorbar(shrink=0.7)
    plt.tick_params(axis='both', left='off', top='off', right='off', bottom='off', labelleft='off', labeltop='off', labelright='off', labelbottom='off')
    plt.subplot(122)
    plt.yscale('log')
    plt.ylim([0.002,0.11])
    plt.title("Combined Average Uncertainty")
    plt.plot(window_values[len(window_values)-100:],'-.',color='black')
    plt.tight_layout()
    plt.tick_params(axis='both', left='on', top='off', right='off', bottom='off', labelleft='on', labeltop='off', labelright='off', labelbottom='off')
    plt.show()
```
